```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# Definir tipos de etiqueta
label_types = ['S', 'M', 'I', 'B', 'H', 'X', 'E']

label_map = {'S': 'stem',
    'M': 'multiloop',
    'I': 'internal loop',
    'B': 'bulge',
    'H': 'hairpin loop',
    'X': 'external loop',
    'E': 'dangling end'}
families = sorted(df['fam'].unique())
cmap = plt.get_cmap('tab10')
color_map = {fam: cmap(i % cmap.N) for i, fam in enumerate(families)}

# Contar etiquetas por secuencia y crear DataFrame
label_counts = df['motivos'].apply(lambda s: Counter(s))
label_counts_df = pd.DataFrame(label_counts.tolist()).fillna(0)[label_types]

# Calcular promedio de cuentas de cada etiqueta por familia
avg_counts = label_counts_df.join(df['fam']).groupby('fam').mean()

# Crear figura con subplots
n_labels = len(label_types)
cols = 4
rows = (n_labels + cols - 1) // cols
fig, axes = plt.subplots(rows, cols, figsize=(16, 8), sharey=True)

# Plot en cada subplot
for idx, label in enumerate(label_types):
    ax = axes[idx // cols, idx % cols]
    data = avg_counts[label].sort_values(ascending=False)
    ax.bar(data.index, data.values, color=[color_map[fam] for fam in data.index])
    ax.set_title(f'{label_map[label]}: ({label})')
    ax.set_xticklabels(data.index, rotation=45, ha='right')
    ax.set_xlabel('Familia')
    ax.set_ylabel('Promedio')
    ax.grid(True)
    
    
# Desactivar subplots vacÃ­os
for idx in range(n_labels, rows * cols):
    fig.delaxes(axes[idx // cols, idx % cols])

plt.tight_layout()
plt.show()
```

```python
import math
import matplotlib.pyplot as plt
from collections import Counter

# --- 1) Definir tipos y mapeos ---
label_types = ['S', 'M', 'I', 'B', 'H', 'X', 'E']
label_map = {
    'S': 'Stem',
    'M': 'Multiloop',
    'I': 'Internal loop',
    'B': 'Bulge',
    'H': 'Hairpin loop',
    'X': 'External loop',
    'E': 'Dangling end'
}

# Colores fijos por motivo
cmap_motives = plt.get_cmap('tab10')
label_colors = {
    lab: cmap_motives(i % cmap_motives.N)
    for i, lab in enumerate(label_types)
}

# --- 2) Calcular fracciÃ³n de cada etiqueta por secuencia ---
label_counts = df['motivos'].apply(lambda s: Counter(s))
label_counts_df = pd.DataFrame(label_counts.tolist()).fillna(0)[label_types]
label_frac_df = label_counts_df.div(df['len'], axis=0)

# --- 3) Agrupar por familia: promedio de fracciÃ³n de cada motivo ---
families = sorted(df['fam'].unique())
avg_frac_by_fam = (
    label_frac_df
    .assign(fam=df['fam'])
    .groupby('fam')
    .mean()
    .loc[families]  # asegurar orden
)

# --- 4) Crear subplots, uno por familia ---
n_fams = len(families)
cols = 4
rows = math.ceil(n_fams / cols)

fig, axes = plt.subplots(rows, cols, figsize=(4*cols, 3*rows), squeeze=False)

for idx, fam in enumerate(families):
    r, c = divmod(idx, cols)
    ax = axes[r][c]
    
    # Obtener y ordenar datos de mayor a menor
    data = avg_frac_by_fam.loc[fam, label_types]
    data_sorted = data.sort_values(ascending=False)
    
    # Dibujar
    ax.bar(
        [label_map[l] for l in data_sorted.index],
        data_sorted.values,
        color=[label_colors[l] for l in data_sorted.index]
    )
    ax.set_title(f'Familia: {fam}')
    ax.set_ylabel('FracciÃ³n media')
    ax.set_xticklabels([label_map[l] for l in data_sorted.index], rotation=45, ha='right')
    ax.grid(axis='y', linestyle='--', alpha=0.5)

# Eliminar ejes sobrantes
for idx in range(n_fams, rows*cols):
    r, c = divmod(idx, cols)
    fig.delaxes(axes[r][c])

fig.suptitle('Motivos ordenados por fracciÃ³n media en cada familia', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# 1) Definir etiquetas y colores
label_types = ['S', 'M', 'I', 'B', 'H', 'X', 'E']
label_map = {
    'S': 'Stem',
    'M': 'Multiloop',
    'I': 'Internal loop',
    'B': 'Bulge',
    'H': 'Hairpin loop',
    'X': 'External loop',
    'E': 'Dangling end'
}
cmap = plt.get_cmap('tab10')
label_colors = {lab: cmap(i % cmap.N) for i, lab in enumerate(label_types)}

# 2) Calcular fracciÃ³n por secuencia
label_counts = df['motivos'].apply(lambda s: Counter(s))
label_counts_df = pd.DataFrame(label_counts.tolist()).fillna(0)[label_types]
label_frac_df = label_counts_df.div(df['len'], axis=0)

# 3) Promedio de fracciÃ³n por familia
avg_frac_by_fam = (
    label_frac_df
    .assign(fam=df['fam'])
    .groupby('fam')
    .mean()
)

# 4) Definir nuevo orden de familias: primero S desc., luego I desc.
fam_order = avg_frac_by_fam.sort_values(['S'], ascending=[False]).index.tolist()

# 5) GrÃ¡fico de barras apiladas con fam_order
fig, ax = plt.subplots(figsize=(12, 6))

bottom = np.zeros(len(fam_order))
for lab in label_types:
    vals = avg_frac_by_fam.loc[fam_order, lab].values
    ax.bar(
        fam_order,
        vals,
        bottom=bottom,
        label=label_map[lab],
        color=label_colors[lab]
    )
    bottom += vals

ax.set_ylabel('FracciÃ³n media')
ax.set_title('DistribuciÃ³n apilada de motivos por familia')
ax.set_xticks(range(len(fam_order)))
ax.set_xticklabels(fam_order, rotation=45, ha='right')
ax.legend(title='Motivo', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
